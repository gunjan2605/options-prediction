```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import seaborn as sns
```

# Read Options Data

```python
#read updated ETF options data
data = pd.read_csv('../data/ETF/ETF_cleaned.csv')
data.drop('Unnamed: 0', axis=1,inplace=True)
data.drop('Unnamed: 0.1', axis=1,inplace=True)
data.drop('index', axis=1,inplace=True)
data=data.reset_index(drop=True)
```

```python
#read updated Tech options data
data_1 = pd.read_csv('../data/Tech_cleaned.csv')
data_1.drop('Unnamed: 0', axis=1,inplace=True)
data_1.drop('Unnamed: 0.1', axis=1,inplace=True)
data_1.drop('index', axis=1,inplace=True)
data_1 = data_1.reset_index(drop=True)
```

```python
data.head(3)
```

```python
data_1.head(3)
```

```python
final_data = pd.concat([data, data_1])
```

```python
final_data = final_data.reset_index(drop=True)
```

```python
final_data
```

# Correlation Graph


Based on the below chart, there are lots of highly correlated predictors (dark red). Leaving highly correlated predictors in the models will lead to bias, so we will take some out. REMOVE - spent, spent_contract, open, high, low, adj close.

 

```python
fig, ax = plt.subplots(figsize=(15,10))      
matrix = np.triu(final_data.corr())
sns.heatmap(final_data.corr(), annot=True, fmt = '.3g',vmin=-1, vmax=1, center= 0, cmap= 'coolwarm', mask=matrix)
```

```python
c = final_data.columns 
corr = final_data[c].corr()
corr.style.background_gradient(cmap='YlOrRd')
```

# Boxplots

```python
plt.figure(figsize =(13, 9))
plt.boxplot([final_data['Open'], final_data['High'], final_data['Low'], final_data['Close'], final_data['Adj Close'], final_data['Strike'], final_data['StockPrice']])
plt.xticks([1, 2, 3,4,5,6,7], ['Open', 'High', 'Low','Close', 'Adj Close','Strike','StockPrice'])
plt.show()
```

```python
predictors = ['Volume','Spent', 'Spent/contract', 'Spent/share', 'break_even', 'Contract period in days', 'attention']

for i in predictors:
  plt.figure(figsize =(22, 2))
  df = pd.DataFrame([final_data[i]], index=[i])
  df.T.boxplot(vert=False)
  plt.yticks(rotation=90)
  plt.show()
```
